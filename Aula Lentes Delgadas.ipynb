# Lentes Convergentes e Divergentes
## Aplicação da equação e Euler

### Faremos um programa interativo que pode ser usado como auxílio para o estudo de formação de imagens de lentes delgadas com uso da equação de Euler: 

# $$\frac{1}{p} + \frac{1}{p'} = \frac{1}{f}$$


## Problemas Possíveis:
>
>### Problemas envolvendo lentes geralmente são divididos em três tipos:
>>
>>### - O valor de p não é dado;
>>### - O valor de p' não é dado;
>>### - O valor de f não é dado.
>
>### Para cada um desses exemplos é possível isolar a variável a ser encontrada em função das outras dadas:
>># $$ p = \frac{p'f}{p' - f} \ ; \ p' = \frac{pf}{p - f} \ ; \ f = \frac{pp'}{p + p'}$$
>
>### A partir disso podemos obter valores do aumento linear da imagem em relação ao objeto:
>># $$\frac{imagem}{objeto} = -\frac{p'}{p}$$
>
>### Para formação da imagem podemos usar os raios notáveis, a partir deles podemos ver onde a imagem se forma pelo encontro de dois desses raios ou pelos prolongamentos dos mesmos. Quando a imagem é formada pela união de dois raios de luz a imagem é dita real (p' >0), quando a imagem é formada por prolongamentos, a imagem é dita virtual (p'<0).
>
>> <img src = "https://www.alfaconnection.pro.br/images/LUZ050201a.gif">
>
>### Também é possível colocar valores muito elevados. Quando o objeto é muito distante da lente, dizemos que ele se encontra no "infinito", e a imagem converge para o foco com tamanho nulo. Quando a imagem se forma muito longe, ou seja, quando p fica muito próximo de f, temos uma imagem imprópia (válida somente para lentes convergentes). Por fim, quando a distância focal é muito elevada, dizemos que a lente se torna plana e pode ser interpretada como um "dioptro plano".
